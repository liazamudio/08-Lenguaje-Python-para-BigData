**Ejercicio 1**

Calcula el área de un triángulo equilátero donde:
*   Base = 4
*   Altura = 6


  

```python
# Declaramos las variables
base = 4
altura = 6

# Calculamos el área
area = (base * altura)/2

# Imprimimos el resultado
print(f'El área de un triangulo equilatero con una base de {base} y una altura de {altura} es de {area}')
```

**Ejercicio 2**

Determina la hipotenusa de un triángulo rectángulo donde los catetos tienen los siguientes valores:
*   A= 4.56
*   B= 5.67

```python
# Declaramos las variables
base = 5.67
altura = 4.56

# Calculamos la hipotenusa
hipotenusa1 = round((math.sqrt((base ** 2) + (altura ** 2))), 2)
# Otra opción es:
# hipotenusa2 = round((((base ** 2) + (altura ** 2)) ** 0.5), 2)

print(f'La hipotenusa de un rectángulo con base {base} y altura {altura} es {hipotenusa1}')
```

**Ejercicio 3**

Establecido el valor de la hipotenusa del ejercicio anterior, ahora calcula el semiperímetro


```python
# Declaramos las variables
base = 5.67
altura = 4.56

# Calculamos la hipotenusa
hipotenusa = round((math.sqrt((base ** 2) + (altura ** 2))), 2)

# Calculamos el semiperimetro
semiperimetro = round(((base + altura + hipotenusa)/2), 2)

print(f'El semiperimetro de un rectángulo con base {base}, altura {altura} e hipotenusa {hipotenusa1} es {semiperimetro}')
```

**Ejercicio 4**

Calcula el seno, coseno y tangente del siguiente triángulo y convierte los valores a grados.

*   X = 12.3
*   Y = 11





```python
# Definimos nuestros valores iniciales
x = 12.3
y = 11

# Determinamos las dimensiones de los tres lados
base = x
altura = y
hipotenusa = math.sqrt(base**2 + altura**2)
# print(f'Hipotenusa: {hipotenusa}')

# Determinamos el seno, coseno y tangente
seno_0 = altura / hipotenusa
coseno_0 = base / hipotenusa
tangente_0 = altura / base

# Determinamos los angulos
# Calcular el ángulo agudo en grados
angulo_agudo = round((math.degrees(math.asin(altura / hipotenusa))), 2)
# Calcular el ángulo obtuso (complemento del ángulo agudo)
angulo_obtuso = 90 - angulo_agudo
# Angulo recto
angulo_recto = 180 - (angulo_obtuso + angulo_agudo)

# Imprimimos resultados:

print(f'Hipotenusa: {hipotenusa}')
print(f'Seno: {seno_0}')
print(f'Coseno: {coseno_0}')
print(f'Tangente: {tangente_0}')
print(f'Ángulo xy: {angulo_agudo} grados.')
print(f'Ángulo 0x: {angulo_obtuso} grados.')
print(f'Ángulo 0y: {angulo_recto} grados.')
```
